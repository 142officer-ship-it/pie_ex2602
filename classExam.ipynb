```python
# 클래스 -> 객체를 표현하기 위한 문법의 일종
# 프로그래밍으로 객체를 만들 때 사용하는 것
# 속성 (명사적)- attribute
# 능력(기능, 동사적) - method, method 매개 변수 자리에 self 있어야 함

class Person:
    def hello(self):
        print("hi")

jin = Person() # jin은 Person의 인스턴스
jin.hello()
```

```python
class Person():
    def __init__(self):
        self.hello = '안녕하세요' # 인스턴스 속성

    def hi (self):
        print(self.hello)

jin = Person() # 이 때 __init__메서드는 이미 실행됐다.
jin.hi()        
```

```python
# class 속성
class Person():
    folder = [] # 클래스 속성 -> 클래스에 속한다(공공재): 모든 인스턴스에서 접근이 가능하다.

    def add_files(self, file):
        self.folder.append(file) # 클래스 이름으로 클래스 속성에 접근할 수 있다.

jin = Person()
jin.add_files('txt')

paul = Person()
jin.add_files('img')

print(jin.folder)
print(paul.folder)
```

```python
class Person():
  def __init__(self):
    self.folder = [] # instance 속성
    
  def add_files(self, file):
    self.folder.append(file)

jin = Person()
jin.add_files('txt')

paul = Person()
jin.add_files('img')

print(jin.folder)
print(paul.folder)
```

```python
class Person:
  def __init__(self):
    self.folder = [] # 인스턴스 속성

  def add_files(self, file):
    self.folder.append(file)

jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
```
