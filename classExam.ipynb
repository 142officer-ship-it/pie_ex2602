```python
# 클래스 -> 객체를 표현하기 위한 문법의 일종
# 프로그래밍으로 객체를 만들 때 사용하는 것
# 속성 (명사적)- attribute
# 능력(기능, 동사적) - method, method 매개 변수 자리에 self 있어야 함

class Person:
    def hello(self):
        print("hi")

jin = Person() # jin은 Person의 인스턴스
jin.hello()
```

```python
# 빈 클래스 만들기
class Person:
    def hello(self):
        print("hello")

    def hi(self):
        self.hello()

jin = Person()
jin.hi()        
```
